```python
import math
import numpy as np
from matplotlib import pyplot as plt


def square_wave(time, period = 2, width=15/60):
    iterPeriod = np.array(period)
    return np.squeeze(np.array([[1 if t % item < width else 0 for t in time] for item in iterPeriod.flat]))

Period = 2
Width = 1

t = np.linspace(0, 10, 2000)
GCD = np.linspace(2.1, 2.5, 2000)
dt = t[1]-t[0]


party_buffs = square_wave(t, period=2, width=15/60)
both = square_wave(t, period=GCD/2.5 + GCD/120, width=15/60 )

bothIntegral = np.sum(party_buffs*both*dt,axis=1)

plt.figure(figsize=(20,12))
plt.plot(GCD,bothIntegral)
plt.show()
```
