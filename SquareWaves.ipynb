```python
import math
import numpy as np
from matplotlib import pyplot as plt


def square_wave(time, period = 2, width=15/60):
    iterPeriod = [period,] if not isinstance(period, (tuple, list)) else period
    return np.squeeze(np.array([[1 if t % item < width else 0 for t in time] for item in iterPeriod]))

Period = 2
Width = 1

t = np.linspace(0, 10, 1000)
GCD = np.linspace(2.17, 2.5, 1000)
dt = t[1]-t[0]


party_buffs = square_wave(t, period=2, width=15/60)
both = square_wave(t, period=2.45/2.5, width=15/60 )

plt.figure(figsize=(20,12))
plt.plot(t,party_buffs*both)
plt.show()
```
